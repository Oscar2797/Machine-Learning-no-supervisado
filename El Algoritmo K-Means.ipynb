```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as mp
import seaborn as sb
import os
os.chdir("C:/Users/odela/Downloads")

from sklearn.cluster import KMeans
```

```python
dataset= pd.read_csv("Mall_customers.csv")
dataset.head()
```

```python
# poner los dos puntos significa del primer renglon al último y ya luego se eligen las columnas
fields= dataset.iloc[:, [3,4]].values
print(fields)
```

```python
import warnings 
warnings.filterwarnings= ("ignore")
```

```python
# Se va a hacer prueba con 11 clusters para encontrar los clusters óptimos
# init es criterio de iniciación y siempre se utiliza en este algoritmo
wcss= []
for index in range(1,11):
    kmeans= KMeans(n_clusters=index, init= "k-means++", random_state= 2)
    kmeans.fit(fields)

    wcss.append(kmeans.inertia_) #.append permite que cada resultado se vaya agregando a una lista, lo cual al final nos va a arrojar los 10 resultados en lista
```

```python
# Graficación del codo de Jambu
sb.set()
mp.plot(range(1,11),wcss) # El eje de las x va a ser el rango de 1 a 11 y y va a ser el resultado de wcss de cada uno de los números del rango
mp.title("K Means clustering")
mp.xlabel("Número de clústeres")
mp.ylabel( "WCSS") # Weighted cluster sum of squares
```

```python
# El codo está en 5, ya que después de eso, la ganancia no es muy grande
```

```python
kmeans= KMeans(n_clusters= 5, init= "k-means++", random_state= 42)
cluster_values= kmeans.fit_predict(fields)
print(cluster_values)
```

```python
mp.figure( figsize=(10,10))
mp.scatter( fields[cluster_values==0,0],fields[cluster_values==0,1], c="blue", label= "Cluster 1") # de los cluster que valen 0, se toma de field la columna 1 com x y la 2 omo y para graficar
mp.scatter( fields[cluster_values==1,0],fields[cluster_values==1,1], c="Orange", label= "Cluster 2")
mp.scatter( fields[cluster_values==2,0],fields[cluster_values==2,1], c="green", label= "Cluster 3")
mp.scatter( fields[cluster_values==3,0],fields[cluster_values==3,1], c="purple", label= "Cluster 4")
mp.scatter( fields[cluster_values==4,0],fields[cluster_values==4,1], c="yellow", label= "Cluster 5")
mp.legend()

mp.scatter(kmeans.cluster_centers_[:,0],kmeans.cluster_centers_[:,1], c="black", label="Centroides") #Graficar centroides
mp.title("Grupo de clientes")
mp.xlabel="Annual income"
mp.ylabel= "Spending score"
```

```python
# Criterio de la silueta ("Silhouette") para conocer los clústeres óptimos

from sklearn import datasets
from sklearn.metrics import silhouette_score
```

```python
X=dataset.iloc[:,[3,4]].to_numpy()
for j in range(2,12):
    kmeans= KMeans(n_clusters=j, random_state=42)
    kmeans.fit_predict(X)
    # Calculamos score de silueta
    score= silhouette_score(X,kmeans.labels_, metric= "euclidean")
    print("Score silhouette", "k=",j,":",score)
```

```python
# Se selecciona el score más cercano a 1, por lo tanto 5 es la respuesta correcta de clusters
```
