```python
import os
os.chdir("C:/Users/odela/Downloads")
import pandas as pd
```

```python
iris1= pd.read_csv("iris.csv")
iris1
```

```python
etiquetas= iris1["variety"]
etiquetas
```

```python
from numpy import array
from sklearn.decomposition import TruncatedSVD
A= iris1
# Remover columna no numérica
A= A.drop(["variety"],axis=1)
A
```

```python
from numpy import diag
from numpy import zeros
from scipy.linalg import svd
import numpy as np
import matplotlib.pyplot as plt
```

```python
U,s, VT= svd(A) # Obtención de matries cuyo producto arroja la matriz A
Sigma= zeros((A.shape[0],A.shape[1])) # creación de matriz de ceros de la misma dimensión que A
Sigma[:A.shape[1],:A.shape[1]]=diag(s)
```

```python
print("Matriz U:")
print(U) # Lo primero es convertir la matriz de 150 renglones y 4 columnaas a 150 renglones y 150 columnas, para eso se usa Matriz U
```

```python
print( "Matriz Sigma")
print(Sigma) # luego va la matriz sigma que vuelve a tener 150 renglones y 4 columnas, tiene muchos ceros
```

```python
print( "Matriz  VT")
print(VT) # Matriz VT
```

```python
# Para reconstruir la matriz original

B= U.dot(Sigma.dot(VT)) # dot es la multiplicación de dos matrices
print(B)
```

```python
# Reducción a 2 dimensiones 

n_elements=2
UReduced2= U[:, :n_elements]
print("U reducida") # U se reduce a la cantida original de renglones y 2 columnas
print(UReduced2)
```

```python
sigma_reduced2= Sigma[:n_elements, :n_elements] # Sigma se debe reducir a 2 renglones y 2 columnas
print(" Sigma reducida")
print(sigma_reduced2)
```

```python
VTreduced2= VT[:n_elements, :]  # VT se debe reducir a 2 renglones y 4 columnas
print("VT reducida")
print(VTreduced2)
```

```python
Areduced2= UReduced2.dot(sigma_reduced2.dot(VTreduced2))
print("Matriz A transformada y reducida")
print(Areduced2) # Puede variar de la base original de iris
```

```python
T2= UReduced2.dot(sigma_reduced2)
print( "Matriz T (2 dimensiones):")
print(T2)
```

```python
df= pd.DataFrame(T2, columns= ["a", "b"])
df.plot(kind="scatter", x="a", y="b")
plt.show()
```

```python
# Reducción a 3 dimensiones
n_elements= 3
```

```python
UReduced3= U[:, :n_elements]


sigma_reduced3= Sigma[:n_elements, :n_elements] # Sigma se debe reducir a 2 renglones y 2 columnas


VTreduced3= VT[:n_elements, :]  # VT se debe reducir a 2 renglones y 4 columnas


Areduced3= UReduced3.dot(sigma_reduced3.dot(VTreduced3))
print("Matriz A transformada y reducida")
print(Areduced3) # Puede variar de la base original de iris
```

```python
T3= UReduced3.dot(sigma_reduced3)
print( "Matriz T (3 dimensiones):")
print(T3)
```
